```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar a base de dados
sinasc_data = pd.read_csv('sinasc_RO_2019.csv')

# Configurar o estilo do seaborn
sns.set(style="whitegrid")
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Criar o boxplot da idade da mãe
plt.figure(figsize=(8, 6))
sns.boxplot(x='IDADEMAE', data=sinasc_data)
plt.title('Boxplot da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# Criar o boxplot do peso do bebê
plt.figure(figsize=(8, 6))
sns.boxplot(x='PESO', data=sinasc_data)
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso do Bebê (gramas)')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Criar o histograma da idade da mãe
plt.figure(figsize=(8, 6))
plt.hist(sinasc_data['IDADEMAE'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
# Filtrar dados para evitar valores nulos no peso do bebê
filtered_data = sinasc_data.dropna(subset=['PESO'])

# Criar o histograma do peso do bebê
plt.figure(figsize=(8, 6))
plt.hist(filtered_data['PESO'], bins=20, color='salmon', edgecolor='black')
plt.title('Histograma do Peso do Bebê')
plt.xlabel('Peso do Bebê (g)')
plt.ylabel('Frequência')
plt.show()
```
