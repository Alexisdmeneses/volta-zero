````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Carregando o DataFrame a partir do arquivo CSV
df = pd.read_csv('previsao_de_renda.csv')

# Exibindo as primeiras linhas do DataFrame
print(df.head())
````

````python
df.dtypes
````

````python
# Gráfico de barras para Posse de Imóvel ao longo do tempo
plt.figure(figsize=(12, 6))
sns.countplot(x='idade', hue='posse_de_imovel', data=df)
plt.title('Distribuição de Posse de Imóvel ao Longo do Tempo')
plt.show()
````

````python
# Gráfico de barras para Posse de Veículo ao longo do tempo
plt.figure(figsize=(12, 6))
sns.countplot(x='idade', hue='posse_de_veiculo', data=df)
plt.title('Distribuição de Posse de Veículo ao Longo do Tempo')
plt.show()
````

````python
# Gráfico de barras empilhadas para médias de Renda por Posse de Imóvel ao longo do tempo
plt.figure(figsize=(12, 6))
sns.barplot(x='idade', y='renda', hue='posse_de_imovel', data=df, ci=None)
plt.title('Médias de Renda por Posse de Imóvel ao Longo do Tempo')
plt.show()
````

````python
# Gráfico de barras empilhadas para médias de Renda por Posse de Veículo ao longo do tempo
plt.figure(figsize=(12, 6))
sns.barplot(x='idade', y='renda', hue='posse_de_veiculo', data=df, ci=None)
plt.title('Médias de Renda por Posse de Veículo ao Longo do Tempo')
plt.show()
````
