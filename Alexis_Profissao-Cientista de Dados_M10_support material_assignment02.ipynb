````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Carregando o DataFrame a partir do arquivo CSV
df = pd.read_csv('previsao_de_renda.csv')

# Exibindo as primeiras linhas do DataFrame
print(df.head())
````

````python
# Criando a matriz de dispersão
sns.set(style="ticks")
sns.pairplot(df[['posse_de_imovel', 'posse_de_veiculo']])
plt.suptitle('Matriz de Dispersão', y=1.02)
plt.show()
````

````python
# Gráfico de dispersão
sns.scatterplot(x='posse_de_imovel', y='renda', data=df)
plt.title('Gráfico de Dispersão: Posse de Imóvel vs. Renda')
plt.show()

sns.scatterplot(x='posse_de_veiculo', y='renda', data=df)
plt.title('Gráfico de Dispersão: Posse de Veículo vs. Renda')
plt.show()
````

````python
# Mapa de calor (usando uma tabela de contingência)
contingency_table = pd.crosstab(df['posse_de_imovel'], df['posse_de_veiculo'])
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues')
plt.title('Mapa de Calor: Posse de Imóvel vs. Posse de Veículo')
plt.show()
````

````python
# Gráfico de Perfis de Médias para Posse de Imóvel
plt.figure(figsize=(12, 6))
sns.pointplot(x='posse_de_imovel', y='renda', data=df, capsize=0.2)
plt.title('Gráfico de Perfis de Médias: Renda por Posse de Imóvel')
plt.show()
````

````python
# Gráfico de Perfis de Médias para Posse de Veículo
plt.figure(figsize=(12, 6))
sns.pointplot(x='posse_de_veiculo', y='renda', data=df, capsize=0.2)
plt.title('Gráfico de Perfis de Médias: Renda por Posse de Veículo')
plt.show()
````
