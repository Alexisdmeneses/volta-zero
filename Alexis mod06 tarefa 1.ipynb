#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar a base de dados
sinasc_data = pd.read_csv('sinasc_RO_2019.csv')

# Exibir os primeiros registros da base de dados
print(sinasc_data.head())
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Contar a frequência de cada valor em APGAR1
apg1_counts = sinasc_data['APGAR1'].value_counts()
```

```python
# Criar gráfico de barras
plt.bar(apg1_counts.index, apg1_counts.values)
plt.title('Distribuição de APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

```python
# Contar a frequência de cada valor em APGAR5
apg5_counts = sinasc_data['APGAR5'].value_counts()

# Criar gráfico de barras
plt.bar(apg5_counts.index, apg5_counts.values)
plt.title('Distribuição de APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

```python
# Contar a frequência de cada valor em SEXO
sexo_counts = sinasc_data['SEXO'].value_counts()

# Criar gráfico de barras
plt.bar(sexo_counts.index, sexo_counts.values, color=['blue', 'pink'])
plt.title('Distribuição de SEXO')
plt.xlabel('SEXO')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

```python
# Contar a frequência de cada valor em GESTACAO
gestacao_counts = sinasc_data['GESTACAO'].value_counts()

# Criar gráfico de barras
plt.bar(gestacao_counts.index, gestacao_counts.values, color='green')
plt.title('Distribuição de GESTACAO')
plt.xlabel('GESTACAO')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

```python
# Contar a frequência de cada valor em ESTCIVMAE
estcivmae_counts = sinasc_data['ESTCIVMAE'].value_counts()

# Criar gráfico de barras
plt.bar(estcivmae_counts.index, estcivmae_counts.values, color='blue')
plt.title('Distribuição de ESTCIVMAE')
plt.xlabel('ESTCIVMAE')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Agrupar por CONSULTAS e calcular a média de APGAR5
avg_apgar5_by_consultas = sinasc_data.groupby('CONSULTAS')['APGAR5'].mean()

# Criar gráfico de barras
avg_apgar5_by_consultas.plot(kind='bar', color='green')
plt.title('Média de APGAR5 por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('Média de APGAR5')

# Mostrar o gráfico
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Configurar o estilo do seaborn para melhorar a estética dos gráficos
sns.set(style="whitegrid")

# Criar um painel de 2x2 com 4 gráficos
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Gráfico 1: Frequência de SEXO
sns.countplot(x='SEXO', data=sinasc_data, ax=axes[0, 0])
axes[0, 0].set_title('Frequência de SEXO')

# Gráfico 2: Frequência de GESTACAO
sns.countplot(x='GESTACAO', data=sinasc_data, ax=axes[0, 1])
axes[0, 1].set_title('Frequência de GESTAÇÃO')

# Gráfico 3: Frequência de ESTCIVMAE
sns.countplot(x='ESTCIVMAE', data=sinasc_data, ax=axes[1, 0])
axes[1, 0].set_title('Frequência de ESTCIVMAE')

# Gráfico 4: Frequência de CONSULTAS
sns.countplot(x='CONSULTAS', data=sinasc_data, ax=axes[1, 1])
axes[1, 1].set_title('Frequência de CONSULTAS')

# Ajustar o layout
plt.tight_layout()

# Mostrar o painel de gráficos
plt.show()
```
